```python
# ----------------------------------------------------------------------
# 1. í•„ìš”í•œ ë¼ì´ë¸ŒëŸ¬ë¦¬ ë° ë°ì´í„° ì¤€ë¹„
# ----------------------------------------------------------------------

import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

# ----------------------------------------------------------------------
# ì˜ˆì‹œ ê±°ë˜ ë°ì´í„° (ì¥ë°”êµ¬ë‹ˆ)
# ----------------------------------------------------------------------
# ê° ë¦¬ìŠ¤íŠ¸ëŠ” í•˜ë‚˜ì˜ ê±°ë˜(Transaction)ë¥¼ ë‚˜íƒ€ëƒ…ë‹ˆë‹¤.
transactions = [
    ['ë§¥ì£¼', 'ì•ˆì£¼', 'ê³¼ì'],
    ['ë§¥ì£¼', 'ì•ˆì£¼', 'ì‚¬ì´ë‹¤'],
    ['ë§¥ì£¼', 'ìš°ìœ '],
    ['ì•ˆì£¼', 'ìš°ìœ ', 'ê³¼ì'],
    ['ë§¥ì£¼', 'ì•ˆì£¼', 'ê³¼ì', 'ìš°ìœ ']
]
```

```python
# ----------------------------------------------------------------------
# ë°ì´í„° ì¤€ë¹„ ë° ì „ì²˜ë¦¬ (ì›-í•« ì¸ì½”ë”©)
# ----------------------------------------------------------------------

# ë°ì´í„°ë¥¼ DataFrame í˜•íƒœë¡œ ë³€í™˜ ë° One-Hot Encoding
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
# ë°ì´í„°ë¥¼ True/False ë§¤íŠ¸ë¦­ìŠ¤ë¡œ ë³€í™˜
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)

print("--- One-Hot Encoding ë°ì´í„° (ì¼ë¶€) ---")
print(df.head())
```

```python
# ----------------------------------------------------------------------
# 1.**ë¹ˆë°œ í•­ëª© ì§‘í•© (Frequent Itemsets) ì°¾ê¸°**
# ----------------------------------------------------------------------
# min_support: ìµœì†Œ ì§€ì§€ë„ (0.6ì€ ì „ì²´ 5ê°œ ê±°ë˜ ì¤‘ 3ê°œ ì´ìƒ í¬í•¨ì„ ì˜ë¯¸)

# **ìµœì†Œ ì§€ì§€ë„ 0.6 (ì „ì²´ ê±°ë˜ 5ê°œ ì¤‘ 3ê°œ ì´ìƒ í¬í•¨) ê¸°ì¤€ìœ¼ë¡œ ë¹ˆë°œ í•­ëª© ì§‘í•© ì°¾ê¸°**
frequent_itemsets = apriori(df, min_support=0.6, use_colnames=True)

print("\n--- ë¹ˆë°œ í•­ëª© ì§‘í•© (Frequent Itemsets) ---")
print(frequent_itemsets)
```

```python
# ----------------------------------------------------------------------
# 2. ì—°ê´€ ê·œì¹™ (Association Rules) ìƒì„± ë° í•„í„°ë§
# ----------------------------------------------------------------------
# metric="confidence": ê·œì¹™ í‰ê°€ ê¸°ì¤€ìœ¼ë¡œ ì‹ ë¢°ë„(confidence) ì‚¬ìš©
# min_threshold=0.75: ìµœì†Œ ì‹ ë¢°ë„ 75% ì´ìƒì¸ ê·œì¹™ë§Œ ì„ íƒ

# ìµœì†Œ ì‹ ë¢°ë„ 0.75 ê¸°ì¤€ìœ¼ë¡œ ì—°ê´€ ê·œì¹™ ìƒì„±
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.75)

print("\n--- ì—°ê´€ ê·œì¹™ (Association Rules) ---")
# 'antecedents' (ì„ í–‰ í•­ëª©) -> 'consequents' (í›„í–‰ í•­ëª©) ìˆœìœ¼ë¡œ ì •ë ¬í•˜ì—¬ ì¶œë ¥
rules.sort_values(by='confidence', ascending=False)
```

```python
# ----------------------------------------------------------------------
# ì—°ê´€ ê·œì¹™ í…Œì´ë¸” ì»¬ëŸ¼ ë° ê°’ í•´ì„
# ----------------------------------------------------------------------

# ê·œì¹™ ì˜ˆì‹œ: (ê³¼ì) -> (ì•ˆì£¼)
# ê°’ ì˜ˆì‹œ: support=0.6, confidence=1.0, lift=1.25

# 1. í•­ëª© ì •ì˜ (Rule Definition)
# 'antecedents': ê·œì¹™ì˜ ì„ í–‰ í•­ëª© (IF ì¡°ê±´). ì˜ˆ: (ê³¼ì)
# 'consequents': ê·œì¹™ì˜ í›„í–‰ í•­ëª© (THEN ê²°ê³¼). ì˜ˆ: (ì•ˆì£¼)

# 2. ê¸°ë³¸ ë¹ˆë„ ë° í™•ë¥  ì§€í‘œ (Frequency & Probability)
# 'support': ê²°í•© ì§€ì§€ë„. P(A âˆ© B). ì „ì²´ ê±°ë˜ ì¤‘ Aì™€ Bê°€ ëª¨ë‘ í¬í•¨ëœ ë¹„ìœ¨.
#            (ì˜ˆ: 0.6ì€ ê³¼ì+ê¸°ì €ê·€ êµ¬ë§¤ ê±°ë˜ê°€ ì „ì²´ì˜ 60%ë¼ëŠ” ì˜ë¯¸)
# 'antecedent support': ì„ í–‰ í•­ëª© Aì˜ ì§€ì§€ë„. P(A).
# 'consequent support': í›„í–‰ í•­ëª© Bì˜ ì§€ì§€ë„. P(B).
# 'confidence': ì‹ ë¢°ë„. P(B|A). Aë¥¼ êµ¬ë§¤í–ˆì„ ë•Œ Bë„ êµ¬ë§¤í•  ì¡°ê±´ë¶€ í™•ë¥ .
#               (ì˜ˆ: 1.0ì€ ê³¼ìë¥¼ ì‚° ê³ ê° 100%ê°€ ê¸°ì €ê·€ë„ ìƒ€ë‹¤ëŠ” ì˜ë¯¸)

# 3. í¥ë¯¸ë„ ë° ê°•ë ¥í•¨ ì§€í‘œ (Interest & Strength)
# 'lift': í–¥ìƒë„. Confidence / Consequent Support. A êµ¬ë§¤ê°€ B êµ¬ë§¤ í™•ë¥ ì„ ë†’ì´ëŠ” ì •ë„.
#         - 1.0ë³´ë‹¤ í¬ë©´: ì–‘ì˜ ì—°ê´€ì„± (í•¨ê»˜ íŒ”ë¦¼). ì˜ˆ: 1.25
#         - 1.0ì´ë©´: ë…ë¦½ì  ê´€ê³„ (ì˜í–¥ ì—†ìŒ).
#         - 1.0ë³´ë‹¤ ì‘ìœ¼ë©´: ìŒì˜ ì—°ê´€ì„± (ì„œë¡œ ëŒ€ì²´ì¬).
# 'conviction': ê·œì¹™ì´ í‹€ë¦´ í™•ë¥ ì˜ ì—­ìˆ˜. ì‹ ë¢°ë„ê°€ 1.0ì¼ ê²½ìš° 'inf' (ë¬´í•œëŒ€).
#               (ì˜ˆ: infëŠ” ê·œì¹™ì´ ë‹¨ í•œ ë²ˆë„ í‹€ë¦¬ì§€ ì•Šì•˜ìŒì„ ì˜ë¯¸)
# 'leverage': ì§€ì§€ë„ì™€ ë…ë¦½ì„±ì˜ ì°¨ì´. P(A âˆ© B) - P(A)P(B).
# 'zhangs_metric': -1ì—ì„œ 1 ì‚¬ì´ë¡œ ì •ê·œí™”ëœ ê·œì¹™ì˜ ê°•ë„ ë° í¥ë¯¸ë„.

# 4. ê¸°íƒ€ ì§€í‘œ
# 'jaccard': A ë˜ëŠ” B ì¤‘ Aì™€ Bê°€ ëª¨ë‘ í¬í•¨ëœ ê±°ë˜ì˜ ë¹„ìœ¨. P(A âˆ© B) / P(A âˆª B).
# 'certainty': ì‹ ë¢°ë„ë¥¼ ë³´ì •í•œ ì§€í‘œ.
# 'kulczynski': P(B|A)ì™€ P(A|B)ì˜ í‰ê· . ê·œì¹™ì˜ ê· í˜• ì¡íŒ ì˜ˆì¸¡ ëŠ¥ë ¥ ì¸¡ì •.
# 'representativity': ì§€ì§€ë„ì™€ ìœ ì‚¬í•˜ë©° ê·œì¹™ì˜ ëŒ€í‘œì„± ì¸¡ì •.
```

```python
# ----------------------------------------------------------------------
# 3. **ê²°ê³¼ ì •ë¦¬** ë° **í•µì‹¬ ì§€í‘œ** í•´ì„
# ----------------------------------------------------------------------
rules_final = rules[[
    'antecedents',     # ì„ í–‰ í•­ëª© (IF A)
    'consequents',     # í›„í–‰ í•­ëª© (THEN B)
    'support',         # ì§€ì§€ë„ (Aì™€ Bê°€ í•¨ê»˜ ë‚˜íƒ€ë‚˜ëŠ” ë¹ˆë„)
    'confidence',      # ì‹ ë¢°ë„ (Aë¥¼ ìƒ€ì„ ë•Œ Bë¥¼ ì‚´ í™•ë¥  P(B|A))
    'lift'             # í–¥ìƒë„ (Aì™€ Bê°€ ìš°ì—° ì´ìƒìœ¼ë¡œ í•¨ê»˜ ë‚˜íƒ€ë‚˜ëŠ” ì •ë„)
]].sort_values(by='confidence', ascending=False)


# ğŸ’¡ ì§€í‘œ í•´ì„ ì˜ˆì‹œ: (ê³¼ì) -> (ì•ˆì£¼) ê·œì¹™
# -------------------------------------------
# support (0.6): ê³¼ìì™€ ì•ˆì£¼ë¥¼ ëª¨ë‘ ì‚° ê±°ë˜ê°€ ì „ì²´ì˜ 60%
# confidence (1.0): ê³¼ìë¥¼ ì‚° ê³ ê°ì€ 100% ê¸°ì €ê·€ë„ í•¨ê»˜ ìƒ€ë‹¤.
# lift (1.25): ê³¼ì êµ¬ë§¤ê°€ ì•ˆì£¼ êµ¬ë§¤ í™•ë¥ ì„ 1.25ë°° ì¦ê°€ì‹œí‚¨ë‹¤.
# - liftê°€ 1.0ë³´ë‹¤ í¬ë©´ (1.25) -> ì–‘ì˜ ì—°ê´€ì„± (í•¨ê»˜ íŒ”ë¦¼)
# - liftê°€ 1.0ì´ë©´ -> ë…ë¦½ì  ê´€ê³„ (ì˜í–¥ ì—†ìŒ)
# - liftê°€ 1.0ë³´ë‹¤ ì‘ìœ¼ë©´ -> ìŒì˜ ì—°ê´€ì„± (ì„œë¡œ ëŒ€ì²´ì¬)

print("\n--- ìµœì¢… ì—°ê´€ ê·œì¹™ (í•µì‹¬ ì§€í‘œ) ---")
print(rules_final)
```
