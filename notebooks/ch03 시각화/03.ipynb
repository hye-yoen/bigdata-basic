```python
import plotly.express as px
import pandas as pd
import numpy as np

#-------------------------------------------------------
# 1.   
#-------------------------------------------------------
np.random.seed(42)
df = pd.DataFrame({
    'Ad_Spend': np.random.uniform(100, 1000, 100),
    'Sales': np.random.uniform(500, 5000, 100),
    'Region': np.random.choice(['East', 'West', 'South'], 100),
    'Month': np.random.choice(['Jan', 'Feb', 'Mar'], 100)
})

#-------------------------------------------------------
# 2. Plotly Express  
#-------------------------------------------------------
fig = px.scatter(
    data_frame=df,               #  DataFrame
    x="Ad_Spend",                # X 
    y="Sales",                   # Y 
    color="Region",              #     ()
    symbol="Month",              #    
    size="Ad_Spend",             #     ()
    size_max=30,                 #   
    hover_name="Region",         #     
    hover_data=["Ad_Spend", "Sales", "Month"],  #    
    text=None,                   #      
    facet_col="Month",           # ()   
    facet_row=None,              #    
    trendline=None,              #   ('ols', 'lowess' )
    trendline_color_override=None,  #   
    opacity=0.7,                 #   (0~1)
    color_discrete_sequence=px.colors.qualitative.Set2,  #  
    title="   (Region )",  #  
    labels={"Ad_Spend": "", "Sales": ""},  #     
    template="plotly_white",     #   ('plotly_dark', 'simple_white' )
    width=900,                   #    ()
    height=500,                  #    ()
    log_x=False,                 # X   
    log_y=False,                 # Y   
    marginal_x="histogram",      # X    ('box', 'violin', 'histogram')
    marginal_y="box",            # Y   
    animation_frame=None,        #    ( )
    animation_group=None,        #   
    range_x=[0, 1200],           # X  
    range_y=[0, 6000],           # Y  
    category_orders={"Region": ["East", "West", "South"]}  #   
)

#-------------------------------------------------------
# 3.  
#-------------------------------------------------------
fig.show()
```

```python
import plotly.express as px
import seaborn as sns

#-------------------------------------------------------
# 1.    (Seaborn  'tips' )
#-------------------------------------------------------
df_tips = sns.load_dataset("tips")

#-------------------------------------------------------
# 2. Plotly Express Box Plot ( )
#-------------------------------------------------------
fig = px.box(
    data_frame=df_tips,           #  
    x="day",                      # X: 
    y="total_bill",               # Y:   
    color="smoker",               #    ()
    points="all",                 #     ('outliers', 'suspectedoutliers', 'all', False)
    notched=True,                 #   notch()     
    boxmode="group",              #    ('group', 'overlay')
    hover_name="sex",             #      
    hover_data=["time", "size"],  #    
    category_orders={"day": ["Thur", "Fri", "Sat", "Sun"]},  # X   
    title="     (  )",  #  
    labels={"total_bill": "   ($)", "day": ""},  # /  
    color_discrete_sequence=px.colors.qualitative.Set2,  #  
    template="plotly_white",      #   ('plotly_dark', 'ggplot2', 'seaborn' )
    width=900,                    #    (px)
    height=500,                   #    (px)
    facet_col="time",             # ()    (/)
    facet_row=None,               #    (None =   )
    animation_frame=None,         #    (  )
    orientation="v",              #   ('v' , 'h' )
    range_y=[0, 60],              # Y  
    log_y=False                   # Y    
)

#-------------------------------------------------------
# 3.  
#-------------------------------------------------------
fig.show()
```

```python
import plotly.express as px
import pandas as pd

#-------------------------------------------------------
# 1.   (  )
#-------------------------------------------------------
# 365    
dates = pd.date_range(start='2024-01-01', periods=365, freq='D')
sales_data = 100 + np.cumsum(np.random.randn(365) * 5)
df_ts = pd.DataFrame({'Date': dates, 'Sales': sales_data})

#-------------------------------------------------------
# 2.     
#-------------------------------------------------------
fig = px.line(
    df_ts, 
    x='Date', 
    y='Sales', 
    title='   (  )'
)

#-------------------------------------------------------
# 3.     ( )
#-------------------------------------------------------
fig.update_xaxes(
    rangeslider_visible=True, #   
    rangeselector=dict(
        buttons=list([
            dict(count=1, label="1m", step="month", stepmode="backward"),
            dict(count=6, label="6m", step="month", stepmode="backward"),
            dict(step="all") #   
        ])
    )
)

#-------------------------------------------------------
# 4.  
#-------------------------------------------------------
fig.show()
```

```python
import plotly.express as px
import pandas as pd
import numpy as np

#-------------------------------------------------------
# 1.   (  )
#-------------------------------------------------------
dates = pd.date_range(start='2024-01-01', periods=365, freq='D')
sales_data = 100 + np.cumsum(np.random.randn(365) * 5)
df_ts = pd.DataFrame({'Date': dates, 'Sales': sales_data, 'Category': np.random.choice(['A', 'B'], 365)})

#-------------------------------------------------------
# 2. Plotly Express Line Chart (  )
#-------------------------------------------------------
fig = px.line(
    data_frame=df_ts,             #  
    x="Date",                     # X: 
    y="Sales",                    # Y:  
    color="Category",             #    ()
    line_group=None,              #     
    line_dash="Category",         #  (/ )  
    symbol=None,                  #      
    markers=True,                 #     (True/False)
    text=None,                    #      
    hover_name="Category",        #      
    hover_data={"Date": True, "Sales": ":.2f"},  #     (:.2f   )
    facet_col=None,               #    
    facet_row=None,               #    
    animation_frame=None,         # ()  
    animation_group=None,         #   
    range_x=["2024-01-01", "2024-12-31"],  # X  
    range_y=[80, 140],            # Y  
    title="   ( )",  #  
    labels={"Date": "", "Sales": ""},  # /  
    template="plotly_white",      #   ('plotly_dark', 'seaborn'  )
    width=900,                    #   (px)
    height=500,                   #   (px)
    log_y=False,                  # Y    
    color_discrete_sequence=px.colors.qualitative.Pastel,  #   
    line_shape="linear"           #    ('linear', 'spline', 'vhv', 'hvh')
)

#-------------------------------------------------------
# 3.  
#-------------------------------------------------------
fig.show()
```
