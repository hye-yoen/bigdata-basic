## OLS

```python
import numpy as np

# 1. 결측치 확인
print(train.isnull().sum())

# 2. 상수 컬럼 확인
print([col for col in train.columns if train[col].nunique() <= 1])

# 3. 상관관계 과도한 컬럼 확인
train.corr(numeric_only=True)


# 4. Inf / NaN 존재 여부
np.isfinite(train.select_dtypes(include=[np.number])).all().all()
```

```python
train.info()
```

```python
#컬럼제거
train.drop('Item_Identifier',axis=1,inplace=True)
test.drop('Item_Identifier',axis=1,inplace=True)
```

```python
#----------------------------------------------
# ★ OLS - 선형회귀에서만 사용할 것(분류는 x)
#----------------------------------------------
from statsmodels.formula.api import ols 

# 항목컬럼명 저장
y_col = 'Item_Outlet_Sales'

# y를 뺀 x값 후보들(수치형,범주형 다 포함됨)
X_cols = [c for c in train.columns if c != y_col]
X_cols

# formula 생성
formula = f"{y_col} ~ " + " + ".join(X_cols) # ols안에 들어갈 문자열 패턴
formula

# OLS모델 적용
model = ols(formula,data=train).fit()

# 확인
print(model.summary())
```
